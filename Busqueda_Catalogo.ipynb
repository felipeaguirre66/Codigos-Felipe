# Busqueda oportunidades catalogo
Este codigo busca en todo mercado libre lo que diga "suplemento en polvo" y lo ordena de mas caro a mas barato

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from lxml import etree
```

```python
siguiente = 'https://listado.mercadolibre.com.ar/suplemento-en-polvo'
lista_titulos = []
lista_precios = []
lista_links = []
actual = 0
final = 0
actual = 0

while actual<=final:
  r = requests.get(siguiente) 

  # Pagina actual
  actual += 1

  if r.status_code == 200:
    soup = BeautifulSoup(r.content, 'html.parser')

    # Titulos
    titulos = soup.find_all('h2', class_ ='ui-search-item__title')
    titulos = [titulo.text for titulo in titulos]
    lista_titulos.extend(titulos)

    # Precios
    dom = etree.HTML(str(soup))
    precios = dom.xpath("//li[@class='ui-search-layout__item']//div[@class='ui-search-price__second-line']//span[@class='price-tag-fraction']")
    precios = [int(i.text.replace('.','')) for i in precios]
    lista_precios.extend(precios)

    # Links
    links = soup.find_all('a', class_ ='ui-search-item__group__element ui-search-link')
    links = [i.get('href') for i in links]
    lista_links.extend(links)

    # Pagina final
    final = int(soup.find_all('li', class_ ='andes-pagination__page-count')[0].text.split(' ')[-1])

  else:
    break
  print(f"Buscando en pagina {actual} de {final}")  
  siguiente = soup.find_all('a', class_ ='andes-pagination__link ui-search-link')[0].get('href')

# Emprolijar
new_lista_titulos = []
new_lista_precios = []
new_lista_links = []

for i,v in enumerate(lista_titulos):
  if 'uplemento en polvo' in v:
    v = v.split('Suplemento en polvo ')[-1]
    new_lista_titulos.append(v)
    new_lista_precios.append(lista_precios[i])
    new_lista_links.append(lista_links[i])
lista_titulos = new_lista_titulos.copy()
lista_precios = new_lista_precios.copy()
lista_links = new_lista_links.copy()
```

```python
# Data Frame
df = pd.DataFrame(np.array([lista_titulos,lista_precios,lista_links]).T, columns=['Producto','Precio','Link'])
df.Precio = df['Precio'].astype(int)
df = df.groupby('Producto')
df = df.first().sort_values('Precio', ascending=False)
df.head(60)
#df.iloc[50:60,:]
```
